#TIME SERIES ANALYSIS

```python
import statsmodels.api as sm
plt.figure(figsize=(15,8))

data_df = pd.read_csv('crime.csv', parse_dates=['YEAR'], index_col=['YEAR'])

# Resample data to yearly frequency
data_yearly = data_df.resample('Y').sum()

# Plotting the time series
plt.plot(data_yearly.index, data_yearly['TOTAL IPC CRIMES'],color='black')
plt.title('Total IPC Crimes over Time')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.show()
#decomposing the data according to the seasons
decomposition = sm.tsa.seasonal_decompose(data_yearly['TOTAL IPC CRIMES'], model='additive')

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 7))
ax1.plot(data_yearly.index, decomposition.observed)
ax1.set_ylabel('Observed')
ax2.plot(data_yearly.index, decomposition.trend,color='red')
ax2.set_ylabel('Trend')
ax3.plot(data_yearly.index, decomposition.seasonal,color='green')
ax3.set_ylabel('Seasonal')
ax4.plot(data_yearly.index, decomposition.resid,color='purple')
ax4.set_ylabel('Residual')
plt.tight_layout()
plt.show()


adf_result = sm.tsa.stattools.adfuller(data_yearly['TOTAL IPC CRIMES'])
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
print('Critical Values:', adf_result[4])
```

```python
min_val = data['TOTAL IPC CRIMES'].min()
max_val = data['TOTAL IPC CRIMES'].max()
range_val = (max_val-min_val)/4

low = min_val + range_val
medium = low + range_val
high = medium + range_val

```

```python
def get_data_level(data_count):
  if data_count <= low:
    return 1
  elif data_count <= medium:
    return 2
  elif data_count <= high:
    return 3
  else:
    return 4

data['DATA_LEVEL']= data['TOTAL IPC CRIMES'].apply(get_data_level)
```

#PROPHET MODEL BEING USED

```python
import pandas as pd
from prophet import Prophet

# Example historical crime rate data
data = pd.DataFrame({
    'ds': pd.date_range(start='2020-01-01', periods=5, freq='Y'),
    'y': [350000, 360000, 370000, 380000, 385000]  # Example crime rates for the years 2020-2024
})

# Base year crime rate
base_year_crime_rate = 380000  # Assuming 2024 as the base year

# Fit Prophet model
model = Prophet()
model.fit(data)

# Make future predictions
future = model.make_future_dataframe(periods=5, freq='Y')  # Predictions for years 2025-2029
forecast = model.predict(future)

# Extract predicted crime rates
predicted_crime_rates = forecast.loc[5:, 'yhat'].values

# Calculate percentage change relative to the base year
percentage_change = ((predicted_crime_rates - base_year_crime_rate) / base_year_crime_rate) * 100

print("Predicted Percentage Change in Crime Rates:")
for year, percentage in zip(range(2025, 2030), percentage_change):
    print(f"Year {year}: {percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt

years = [2025, 2026, 2027, 2028, 2029]
percentage_change = [3.97, 6.61, 9.24, 10.56, 13.22]
plt.figure(figsize=(10, 6))
plt.plot(years, percentage_change, marker='o', linestyle='-', color='r', label='Percentage Change')
plt.title('Predicted Percentage Change in Crime Rates (2025-2029)')
plt.xlabel('Year')
plt.ylabel('Percentage Change (%)')
plt.xticks(years)
plt.grid(True)# to darw the grid in graph
plt.legend()
plt.tight_layout()

for year, change in zip(years, percentage_change):
    plt.text(year, change, f'{change:.2f}%', ha='right', va='bottom', fontsize=10)

# Show plot
plt.show()
```
